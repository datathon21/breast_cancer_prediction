```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.drop('id',axis=1)
```

```python
df.describe()
```

```python
def parse_diagnosis(diagnosis):
    if diagnosis == 'B':
        return 0
    else:
        return 1
```

```python
df['diagnosis'] = df['diagnosis'].apply(parse_diagnosis)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Unnamed: 32', axis=1)
```

```python
df.info()
```

```python
plt.figure(figsize=(8,4),dpi=200)
plt.ylabel('Texture Mean')
plt.xlabel('Radius Mean')
sns.scatterplot(x='radius_mean', y='texture_mean',data=df,hue='diagnosis',alpha=0.7)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('diagnosis',axis=1)
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(scaled_X_train, y_train)
```

```python
log_model.coef_
```

```python
y_pred = log_model.predict_proba(scaled_X_test)
```

```python
y_pred
```

```python
log_model.coef_
```

```python
df.head()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred = log_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(log_model,scaled_X_test,y_test,normalize='all')
```

```python
len(y_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import plot_precision_recall_curve, plot_roc_curve
```

```python
fig,ax = plt.subplots(figsize=(10,5),dpi=200)
plot_roc_curve(log_model, scaled_X_test, y_test,ax=ax)
```

```python
plot_precision_recall_curve(log_model, scaled_X_test, y_test)
```
